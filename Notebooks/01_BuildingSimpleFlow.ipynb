## Building a simple flow using langgraph

```python
from pydantic import BaseModel, Field
from langgraph.graph import StateGraph,START,END
from IPython.display import Image, display
from typing import Literal
import random
```

```python
#Define State
class State(BaseModel):
    text: str = Field(..., description="The text content of the state")
```

```python
# Define Nodes
def start_play(state:State)-> State:
    print("start node running...")
    return State(text=state.text + "I'm planning to play")
def cricket(state: State)-> State:
    print("cricket node running...")
    return State(text=state.text + " cricket")

def football(state: State)-> State:
    print("football node running...")
    return State(text=state.text + " football")

def random_play(state:State)-> Literal["cricket", "football"]:
    graph_info=state.text
    if random.random()>0.5:
        return "cricket"
    else:
        return "football"
```

```python
#Build Graph
graph=StateGraph(State)

# Adding Nodes to the graph
graph.add_node("start_play",start_play)
graph.add_node("cricket",cricket)
graph.add_node("football",football)


# Adding Edges to the graph
graph.add_edge(START,"start_play")
#graph.add_edge("cricket","decision_node")
graph.add_conditional_edges("start_play",random_play)
# graph.add_edge("decision_node","cricket")
# graph.add_edge("decision_node","football")
graph.add_edge("cricket",END)
graph.add_edge("football",END)
```

```python
#  Compile the graph
compiled_graph = graph.compile()
```

### Graph Invocation

```python
compiled_graph.invoke({"text":"Hi My name is Raj,"})
```
